# Basic regression tests

```python
import warnings
warnings.filterwarnings('ignore')

from model_tools.regression import *
from model_tools.utils import *
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Single variable linear function without noise.

The closed form of the function is $y=ax+b$.

### Test 1 : $a=1$, $b=0$.

```python
x = np.array(range(100))
y = np.array(range(100))
data1 = pd.DataFrame({'x':x.tolist(), 'y':y.tolist()})
data1.head()
```

```python
from sklearn.linear_model import LinearRegression
reg = Regressor(LinearRegression(), "lr")
data = Data(data1, ['x'], 'y', 0.8)
setting = Setting(transforming_strategy="")
p1 = CustomPipeline(reg, data, setting)
p1.get_summary()
```

```python
predictions = p1.get_prediction()

plt.scatter(data=data1, x='x', y='y', s=5, c='blue')
plt.scatter(x=p1.data.X_test, y=predictions, s=5, c='red')
plt.show()
```
